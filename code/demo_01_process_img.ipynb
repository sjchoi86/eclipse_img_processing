# Basic Image Process

```python
import cv2,glob,os
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline 
# %config InlineBackend.figure_format = 'retina'
print ("Packaged loaded.")
```

### Get Image Paths

```python
img_paths = [f for f in sorted(glob.glob("../raw_img/ROH_*.JPG"))]
n_img = len(img_paths)
print ("We have [%d] images."%(n_img))
```

### Find Circles inside the Image using Hough Transform

```python
FIRST_FLAG,n_plot = True,100
for it,idx in enumerate(np.round(np.linspace(0,n_img-1,n_plot)).astype(np.int)):
    # Load
    img_path = img_paths[idx]
    img_raw_bgr = cv2.imread(img_path)
    img_raw_rgb = cv2.cvtColor(img_raw_bgr,cv2.COLOR_BGR2RGB) # RGB
    scale = 0.2
    width,height = int(img_raw_rgb.shape[1]*scale),int(img_raw_rgb.shape[0]*scale)
    img_rsz_rgb = cv2.resize(img_raw_rgb,(width,height))
    img_raw_gray = cv2.cvtColor(img_raw_bgr, cv2.COLOR_BGR2GRAY)
    img_rsz_gray = cv2.cvtColor(img_rsz_rgb, cv2.COLOR_BGR2GRAY)

    # Find a circle and Draw Green Circles on the Original Image 
    circles = cv2.HoughCircles(cv2.medianBlur(img_rsz_gray,ksize=7),
                               method=cv2.HOUGH_GRADIENT,dp=1,minDist=20,
                               param1=50,param2=5,minRadius=30,maxRadius=50)
    img_raw_rgb_w_circle = np.copy(img_raw_rgb)
    if circles is None: circles = []
    for circle in circles:
        c,r = (int(circle[0,0]/scale),int(circle[0,1]/scale)),circle[0,2]/scale
        r = int(r + 2.0)
        cv2.circle(img_raw_rgb_w_circle,center=c,radius=r,color=(0,255,0),thickness=20)
        cv2.circle(img_raw_rgb_w_circle,center=c,radius=20,color=(0,0,255),thickness=20)
        
    # Mask
    height,width = img_raw_rgb.shape[0],img_raw_rgb.shape[1]
    mask = np.zeros((height,width), np.uint8)
    cv2.circle(mask,center=c,radius=int(r+10),color=(255,255,255),thickness=-1)
    img_mask = cv2.bitwise_and(img_raw_rgb,img_raw_rgb,mask=mask)
    
    # Crop
    _,thresh = cv2.threshold(mask,1,255,cv2.THRESH_BINARY)
    contours = cv2.findContours(thresh,cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)
    x,y,w,h = cv2.boundingRect(contours[0])
    if FIRST_FLAG:
        FIRST_FLAG = False
        crop_xy,crop_wh = (x,y),(w,h)
    img_crop = cv2.resize(img_mask[y:y+h,x:x+w],dsize=crop_wh) 
    
    # Additional Masking to the Cropped Image
    img_crop_gray = cv2.cvtColor(img_crop, cv2.COLOR_BGR2GRAY)
    gray_th = 30
    _,img_crop_gray_th = cv2.threshold(img_crop_gray, gray_th, 255, cv2.THRESH_TOZERO)
    img_crop_mask = np.copy(img_crop)
    img_crop_mask[img_crop_gray_th==0] = 0
    
    # Histogram eq (not helping that much so we don't use this one.)
    r,g,b = cv2.split(img_crop_mask)
    r = cv2.equalizeHist(r)
    img_crop_mask_eq = cv2.merge([r,g,b])
    
    # Make the post-processed image by adding cropped sun into a black image
    img_post = np.zeros_like(img_raw_rgb)
    x,y,w,h = crop_xy[0],crop_xy[1],crop_wh[0],crop_wh[1]
    img_post[y:y+h,x:x+w] = img_crop_mask
    
    # Save the post-processed image
    raw_name = os.path.splitext(os.path.split(img_path)[1])[0]
    save_path = '../post_img/' + raw_name + '_post.jpg'
    cv2.imwrite(save_path, cv2.cvtColor(img_post,cv2.COLOR_RGB2BGR))
    print ("[%s] saved."%(save_path))
    
    # Plot 
    if (it % 10) == 0:
        plt.figure(figsize=(20,3))
        plt.subplot(1,6,1); plt.imshow(img_raw_rgb_w_circle); plt.title("[%d/%d] %s"%(idx,n_img,img_path),fontsize=10)
        plt.subplot(1,6,2); plt.imshow(img_mask); plt.title("Masked",fontsize=10)
        plt.subplot(1,6,3); plt.imshow(img_crop); plt.title("Cropped",fontsize=10)
        plt.subplot(1,6,4); plt.imshow(img_crop_mask); plt.title("Cropped + Masked",fontsize=10)
        plt.subplot(1,6,5); plt.imshow(img_crop_mask_eq); plt.title("Cropped + Masked + Eq",fontsize=10)
        plt.subplot(1,6,6); plt.imshow(img_post); plt.title("Post-processed Image",fontsize=10)
        plt.show()
    
print ("Done.")
```

### Make video

```python
img_paths = [f for f in sorted(glob.glob("../post_img/*.jpg"))]
img_array = []
for idx in range(len(img_paths)):
    filename = img_paths[idx]
    img = cv2.imread(filename)
    height, width, layers = img.shape
    size = (width,height)
    img_array.append(img)

out = cv2.VideoWriter('../vid/seq.avi',cv2.VideoWriter_fourcc(*'DIVX'), 10, size)
for i in range(len(img_array)):
    out.write(img_array[i])
out.release()
print ("Done.")
```
